Import the data and change the list on a DataFrame

```python
import pandas as pd
d = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M', header = 0)
df = d[0]
```

Ignore the cells with a borough that is Not assigned

```python
df_filtrer = df.drop(df[df['Borough']=="Not assigned"].index,axis=0)
```

Groupe the same value and Combined the rows with the neigborhoods separated with a comma

```python
df_drop = df_filtrer.groupby("Postcode").agg(lambda x:','.join(set(x)))
```

```python
df_drop.loc[df_drop['Neighbourhood']=='Not Assigned', 'Neighbourhood']=df_drop.loc[df_drop['Neighbourhood']=='Not Assigned', 'Borough']
```

```python
geo_data = pd.read_csv('http://cocl.us/Geospatial_data')
```

```python
df_drop['Latitude'] = geo_data['Latitude'].values
df_drop['Longitude'] = geo_data['Longitude'].values
```

```python
df_drop
```

```python
import folium
# create map of Toronto using mean latitude and mean longitude values
map_toronto = folium.Map(location=[df_drop.Latitude.mean(), df_drop.Longitude.mean()], zoom_start=10)

# add markers to map
for lat, lng, label in zip(df_drop['Latitude'], df_drop['Longitude'], df_drop['Neighbourhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)

map_toronto
```
